```python
#Загрузим необходимые библиотеки
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Исходные данные: x - расходы на рекламу, Y - объемы продаж
x = np.array([3, 5, 7, 6, 9])
Y = np.array([3, 7, 11, 14, 15])
display(x, Y)
```

```python
#Изобразим их на графике
plt.figure()
plt.scatter(x, Y)
plt.show()
```

```python
#Вычислим оценки коэффициентов парной линейной регрессии по формулам
a1 = ((x - x.mean())*(Y - Y.mean())).mean()/((x - x.mean())**2).mean()
a0 = Y.mean() - a1*x.mean()
print("Модель линейной регрессии: Y^ = ", a0, " + ", a1, "* x")
```

```python
#Дадим серию прогнозов Y^ для x от 3 до 9 с шагом 1
x_space = np.linspace(3, 9, 7)
print(x_space)
Y_pred = a0 + a1*x_space
print(Y_pred)
```

```python
#Реализуем цикл градиентного спуска
class SimpleRegression(object):
    def __init__(self):
        self.a0 = 0
        self.a1 = 0
    def predict(self, x):
        return self.a0 + self.a1*x
    def MSE(self, x, Y):
        return ((Y - self.predict(x))**2).mean()
    def MAE(self, x, Y):
        return abs(Y - self.predict(x)).mean()
    def MAPE(self, x, Y):
        return abs((Y - self.predict(x))/Y).mean()
    def fit(self, x, Y, alpha = 0.001, epsylon = 0.01, max_steps = 5000):
        steps, errors = [], []
        step = 0
        for _ in range(max_steps):
            dT_a0 = -2*sum((Y -self.predict(x)))
            dT_a1 = -2*sum((Y -self.predict(x))*x)
            self.a0 -= alpha*dT_a0
            self.a1 -= alpha*dT_a1
            new_error = self.MSE(x, Y)
            step += 1
            steps.append(step)
            errors.append(new_error)
            if new_error < epsylon:
                break
        return steps, errors
```

```python
#Запустим цикл градиентного спуска с заданной точностью 5
regr = SimpleRegression()
steps, errors = regr.fit(x, Y, alpha = 0.001, epsylon = 5)
```

Загрузите файл Гиперспектр кукурузы.csv (с помощью pd.read_csv).

1. Вычислите аналитическим путем оценки коэффициентов парной линейной регрессии и постройте ее график.
2. Найдите оценки параметров модели парной лиенйной регрессии с использованием метода градиентного спуска. Постройте график.
3. Оцените качество построенных моделей сравнив на графике обучающую выборку и прогнозы.


```python
#1. Загрузка данных и предварительная обработка
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Загрузка данных (учитываем, что в файле используется ';' как разделитель и ',' как десятичный разделитель)
data = pd.read_csv('Гиперспектр кукурузы.csv', sep=';', decimal=',')
data.columns = ['wavelength', 'spectrum']  # Переименуем колонки для удобства

# Выведем первые 5 строк для проверки
print(data.head())
```

```python
#2. Аналитическое решение для парной линейной регрессии
# Вычислим коэффициенты аналитически
x = data['wavelength']
y = data['spectrum']

beta_1 = np.sum((x - x.mean()) * (y - y.mean())) / np.sum((x - x.mean())**2)
beta_0 = y.mean() - beta_1 * x.mean()

print(f"Аналитические коэффициенты: beta_0 = {beta_0:.4f}, beta_1 = {beta_1:.6f}")
```

```python
#3. График аналитической регрессии
plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=5, label='Данные', color='blue')
plt.plot(x, beta_0 + beta_1 * x, label=f'Аналитическая регрессия: y = {beta_0:.4f} + {beta_1:.6f}x', color='red')
plt.xlabel('Длина волны (нм)')
plt.ylabel('Интенсивность спектра')
plt.title('Гиперспектр кукурузы и линейная регрессия')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#4. Решение методом градиентного спуска
# Инициализация параметров
learning_rate = 0.00000192
epochs = 1000
beta_0_gd = 0
beta_1_gd = 0
n = len(x)

# Градиентный спуск
for _ in range(epochs):
    y_pred = beta_0_gd + beta_1_gd * x
    d_beta_0 = (-2/n) * np.sum(y - y_pred)
    d_beta_1 = (-2/n) * np.sum((y - y_pred) * x)
    beta_0_gd -= learning_rate * d_beta_0
    beta_1_gd -= learning_rate * d_beta_1

print(f"Градиентный спуск: beta_0 = {beta_0_gd:.4f}, beta_1 = {beta_1_gd:.6f}")
```

```python
#5. Сравнение моделей
plt.figure(figsize=(12, 6))
plt.scatter(x, y, s=5, label='Данные', color='blue')
plt.plot(x, beta_0 + beta_1 * x, label='Аналитическая', color='red', linestyle='--')
plt.plot(x, beta_0_gd + beta_1_gd * x, label='Градиентный спуск', color='green', linestyle='-.')
plt.xlabel('Длина волны (нм)')
plt.ylabel('Интенсивность спектра')
plt.title('Сравнение методов регрессии')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#6. Оценка качества моделей
# Функция для вычисления MSE
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

# Оценка ошибок
y_pred_analytical = beta_0 + beta_1 * x
y_pred_gd = beta_0_gd + beta_1_gd * x

print(f"MSE аналитическая: {mse(y, y_pred_analytical):.6f}")
print(f"MSE градиентный спуск: {mse(y, y_pred_gd):.6f}")
```
